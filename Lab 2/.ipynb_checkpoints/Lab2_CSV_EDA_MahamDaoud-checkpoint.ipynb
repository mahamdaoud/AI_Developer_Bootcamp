```python
''' Loading essential Liabraries'''
import pandas as pd
import numpy as np
''' loading dataset '''
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
```

```python
#reading the data and storing into a variable named as df
df = pd.read_csv(url)
#cheking the shape of data
print("Shape:", df.shape)
```

#### In the above example, 25% is first Quartile --> Q1 --> (means 25% data value less than or equal to this number)
- 50% means Second Quartile --> Q2 --> (means 50% od data values are above the value and 50% of values are below this value)
- 75% means third quartile --> Q3 --> (means 75% of data values are below to equal to this value)
- IQR --> Inter Quartile range --> (Q3 - Q1)

```python
df.isnull().sum() # adding all the nul values in the dataset
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True) #filling the null valuse using median becaue age has some high and very low values that's why we will use median.
#inplace true meaning we will change the value in the data itself. Direct change in the original data.
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace = True) #drop duplicate values in the original data
```

```python
#importing liabraries for visualizing the data
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style ='whitegrid') #default visual style for all plots that will come after this.
```

```python
sns.countplot(x='Sex', data=df)
```

##### sns.countplot() --> seaborn function for counting categorical values
- x = 'Sex' --> The categroical column to count
- data = df --> the dataframe that contains the data

```python
plt.xlabel("Gender")
plt.ylabel("Values")
plt.title("Gender Distribution")
```

```python
plt.show()
```

When we used all the lines of code seperately only first showed the result. rest showed nothing.

```python
sns.countplot(x = 'Sex', data = df)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Distribution")
plt.show()
```

```python
correlation = df.corr(numeric_only = True)
sns.heatmap(correlation, annot = True, cmap ='coolwarm')
plt.title("Feature Correlation HeatMap")
plt.show()
```
